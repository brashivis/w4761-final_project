### When we download and decompress the database, there is a folder "all_ct_files" where it contains every RNA secondary structure. It is of .ct notation which is difficult for our application. Hence we need to convert it to dot-parenthesis format using a variety of tools.

# Build a list of all .ct files within the folder "all_ct_files"

```python
from os import listdir
from os.path import isfile, join
#modify the path as needed
onlyfiles = [f for f in listdir("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files") if isfile(join("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files", f))]
print (len(onlyfiles))
```

# The .ct files contains hashtags in the file blocking conversion, further processing is need to remove the unnecessary lines

```python
#modify the path as needed
#Read and write the file again without hashtag
for index in range(len(onlyfiles)):
    o=open("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files/tmp/" + onlyfiles[index],"a")
    f=open("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files/" + onlyfiles[index])
    for line in f:
        if "#" not in line:
            o.write(line)
    f.close()
    o.close()
```

## Print out all file names to be copied into excel to produce command line commands to convert .ct files into dot parenthesis command. The finished command line file is stored as "ct2dot_cmd.txt" file. Put the file together with the cleaned .ct files

# After successful conversion, the dot files are moved into /dot folder and loaded back to notebook to convert to .csv file

```python
dotfiles = [f for f in listdir("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files/dot") if isfile(join("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files/dot", f))]
print (len(dotfiles))
```

# Integrate the .dot files into a pandas dataframe

```python
pandas_list = []
for file in dotfiles:
    f = open("C:/Users/KEVIN/RNA_STRAND_data/all_ct_files/dot/" + file)
    list_index = 0
    row = ["","",""]
    row[list_index] = file[:-4]
    list_index += 1
    for line in f:
        if list_index > 2:
            continue
        if ("energy" not in line.lower()):
            row[list_index] = line.rstrip()
            list_index += 1
    pandas_list.append(row)
```

```python
import pandas as pd
df = pd.DataFrame(pandas_list, columns =['RNA_name', 'sequence', 'secondary_structure']) 
df 
```

# Create two more columns to check the conversion is successful that both sequence and secondary structure match in length

```python
df['sequence_length'] = df.apply(lambda x: len(x.sequence), axis=1)
df['secondary_structure_length'] = df.apply(lambda x: len(x.secondary_structure), axis=1)
df
```

```python
#Saving the finished dataset
df.to_csv('RNA_secondary_structure.csv', index = False)
```
